## Spearman Correlation

Spearman Correlation evaluates the monotonic relationship


To compute spearman's correlation we need to compute rank of each value

* Spearman's correlation is little higher than pearson correlation
    * If relationship between distributions are non linear, spearman's correlation tends to better estimate the strength of relationship
    * Pearson correlation can be affected by outliers. Spearman's correlation is more robust.

## Q4. What are the advantages of Spearman Correlation over Pearson Correlation?


The advantages of Spearman Correlation over Pearson Correlation is:-
    
1. Spearman capture non-linear property while pearson only capture linear property
2.Spearman correlation evaluates the monotonic relationship
3.The Spearman Correlation cofficient is based on the ranked values for each variable rather than the raw data.
